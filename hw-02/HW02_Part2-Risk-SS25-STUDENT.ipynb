### Graph Creation

&#9989; **Question 3.2 (7 points)**: Create a NetworkX graph object from the territory definitions. The territories are the nodes of the graph. Implement a function that builds the graph and adds appropriate attributes for continents. _Hint:_ Look up the documentation of `add_node`, what are the possible inputs? Don't forget to connect each territory to its neighbors. Make sure to add plenty of comments in your code.

```python
import networkx as nx

def create_risk_graph(territories):
    """
    Create a NetworkX graph from the Risk territory definitions.

    Parameters:
        territories (dict): Dictionary containing territory and connection information,
                            organized by continent.

    Returns:
        G (nx.Graph): NetworkX graph representing the Risk board, with territories as nodes
                      and 'continent' as a node attribute.
    """
    # Initialize an undirected graph
    G = nx.Graph()

    # Loop through each continent and its territories
    for continent, continent_territories in territories.items():
        for territory, neighbors in continent_territories.items():
            # Add the territory as a node and store its continent as an attribute
            G.add_node(territory, continent=continent)

            # Add an edge between this territory and each of its neighbors
            for neighbor in neighbors:
                G.add_edge(territory, neighbor)

    return G

# Create the graph
G = create_risk_graph(risk_territories)

print(G.nodes['Egypt'])
# Output: {'continent': 'Africa'}

# View all neighbors of a territory
print(list(G.neighbors('Egypt')))
```

### Graph Visualization

&#9989; **Question 3.3 (4 points)**: Create a visualization of the Risk board graph. You can use AI to help you with this. The function below should:

- make a plot of the risk graph.
- color nodes by continent
- allow the user to choose a type of layout. _Hint:_ search networkx documentation for the possible layout. Min 5 layouts should be available.
- print the name of each territory on the respective node
- make a legend for the graph 
- have very descriptive comments 
- increase the size of each node from the default value  
- increase the font size of label from the default value

```python
import matplotlib.pyplot as plt
import networkx as nx
from matplotlib.patches import Patch

def visualize_risk_graph(G, layout_type='spring'):
    """
    Create a visualization of the Risk board graph with continents colored differently.

    Parameters:
        G (nx.Graph): NetworkX graph of the Risk board
        layout_type (str): Layout to use for visualization. Options: 
                           'spring', 'circular', 'kamada_kawai', 'random', 'shell'
    """
    # Define layout options and get the function for selected layout
    layout_functions = {
        'spring': nx.spring_layout,
        'circular': nx.circular_layout,
        'kamada_kawai': nx.kamada_kawai_layout,
        'random': nx.random_layout,
        'shell': nx.shell_layout
    }

    if layout_type not in layout_functions:
        print(f"Invalid layout '{layout_type}'. Using default 'spring' layout.")
        layout_type = 'spring'

    pos = layout_functions[layout_type](G)

    # Define colors for each continent
    continent_colors = {
        'North America': 'skyblue',
        'South America': 'orange',
        'Europe': 'violet',
        'Africa': 'lightgreen',
        'Asia': 'gold',
        'Australia': 'coral'
    }

    # Assign a color to each node based on its continent
    node_colors = []
    for node in G.nodes():
        continent = G.nodes[node].get('continent', 'Unknown')
        node_colors.append(continent_colors.get(continent, 'gray'))

    # Start plotting
    plt.figure(figsize=(16, 12))

    # Draw nodes with increased size and color
    nx.draw_networkx_nodes(
        G, pos,
        node_size=1000,
        node_color=node_colors,
        edgecolors='black'
    )

    # Draw edges between territories
    nx.draw_networkx_edges(G, pos, width=1.5)

    # Draw territory names with larger font size
    nx.draw_networkx_labels(
        G, pos,
        font_size=10,
        font_color='black'
    )

    # Create a custom legend using matplotlib patches
    legend_elements = [
        Patch(facecolor=color, edgecolor='black', label=continent)
        for continent, color in continent_colors.items()
    ]

    plt.legend(
        handles=legend_elements,
        title="Continents",
        loc='best',
        fontsize=10,
        title_fontsize=12
    )

    # Remove axes and show plot
    plt.axis('off')
    plt.title("Risk Board Graph", fontsize=16)
    plt.tight_layout()
    plt.show()
visualize_risk_graph(G, layout_type='kamada_kawai')

```
